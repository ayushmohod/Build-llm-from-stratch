````python
#try to do the comparisonX
torch_rand1=torch.rand(10000,10000).to(device)#create 10000X10000 matrix and give to gpu(to(device))
torch_rand2=torch.rand(10000,10000).to(device)
np_rand1=torch.rand(10000,10000)#create 10000X10000 matrix and give to cpu
np_rand2=torch.rand(10000,10000)
#can use %%time it will tell how long does a block take to execute, cpu time is time for operation(execution)
#and wall time(execution + waiting) is how long we have to wait until it finished
````

````python
%%time
#using torch and cuda

start_time1=time.time()
rand=(torch_rand1 @ torch_rand2)#to multiply in torch we use "@" here
end_time1=time.time()

elapsed_time=end_time1-start_time1
print(f"{elapsed_time:10f}")
````

````python
%%time
#using numpy 

start_time2=time.time()
rand=np.multiply(np_rand1,np_rand2)
end_time2=time.time()

elapsed_time=end_time2-start_time2
print(f"{elapsed_time:10f}")
````

````python
#see a big difference time
#torch.multinomial
prob=torch.tensor([0.1,0.9])# define probabilty tensor the must add up to 1 i.e. 100%, 0.1 means 10% we get index 0 and 0.9
# means 90% chance that we get a index 1 i.e. why in out we see lot of 1's
samples=torch.multinomial(prob,num_samples=10, replacement=True)
# While a binomial distribution describes the outcomes of binary scenarios (e.g., flipping a coin),
# a multinomial distribution applies to cases where there are more than two possible outcomes.
# For instance, rolling a die has six possible outcomes.
# replacement=True means that after an item is selected, it's "put back into the pool," and it could be selected 
# again in the same sampling operation. This is akin to drawing balls from a bag, then putting the ball back each time you draw,
# allowing for the possibility of drawing the same ball multiple times.
print(samples)
````

````python
out=torch.zeros(5,5).masked_fill(torch.tril(torch.ones(5,5))==0,float('-inf'))
# torch.tril(torch.ones(5,5))==0: This comparison checks which elements in the lower triangular matrix are equal to 0, 
# resulting in a Boolean tensor of the same size. Elements that are 0 (above the diagonal) become True, and elements 
# that are 1 (on and below the diagonal) become False.
# torch.zeros(5,5).masked_fill(..., float('-inf')): The masked_fill operation takes a mask (the Boolean tensor from the previous 
# step) and a value. It replaces all positions in the original tensor (the 5x5 tensor of zeros) where the mask is True with
# the specified value, in this case, float('-inf'), which represents negative infinity. Since the mask is True for 
# elements above the diagonal (where the original condition was ==0), those positions in the zeros tensor get replaced with 
# negative infinity. The elements on and below the diagonal, where the mask is False, remain unchanged (i.e., 0).
out
````

````python
#[1,2,3]->softmax, what softmax does is that it exponentiate each number and divide it by the sum of all,
#like for exp(1)=2.71, exp(2)=7.34, exp(3)=19.9, total sum=29.95
# [2.71/29.95,7.34/29.95,19.9/29.95]==[0.0900.0.2447,0.6652]
# Probability Distribution:
# The softmax function converts a vector of raw scores (also called logits) into a probability distribution. 
# The output probabilities are proportional to the exponentials of the input numbers. This is especially useful in multi-class 
# classification tasks where the model needs to give the probability that the input belongs to each class.
````

````python
import torch.nn as nn

# Creating an embedding layer
embedding_layer = nn.Embedding(num_embeddings=80, embedding_dim=6)

# Assuming each number corresponds to a word in the vocabulary
input_ids = torch.tensor([0, 2, 4, 7])  # Indices for "great", "delicious", "friendly", "hated"

# Getting the embeddings for our input_ids
embeddings = embedding_layer(input_ids)
print(embeddings.size())
print(embeddings)
````
